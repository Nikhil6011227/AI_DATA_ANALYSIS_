```python
# Common Data Errors Examples

# 1. Missing Data:
# Task 1: Review a dataset where some customer emails are missing. Identify how
# many records are incomplete.
# Task 2: Examine a sales dataset with missing transaction dates and determine the
# percentage of missing data.
# Task 3: Identify missing department information in an employee registry.






# 2. Duplicate Data:
# Task 1: Analyze a customer dataset with duplicate entries and count the number of
# duplicates.
# Task 2: Review supplier data and identify any repeated supplier names.
# Task 3: Examine a product inventory list for duplicates in product IDs.






# 3. Inconsistent Formatting:
# Task 1: Spot inconsistencies in date formats (e.g., DD/MM/YYYY vs. MM/DD/YYYY)
# in a dataset.
# Task 2: Identify phone numbers with varying formats in a contact list.
# Task 3: Review address data for discrepancies in state abbreviations (e.g., CA vs.
# Calif.).





# 4. Data Drift:
# Task 1: Compare monthly revenues over six months to identify data drift.
# Task 2: Analyze user engagement metrics from a web application over different
# quarters.
# Task 3: Review a stock price dataset to detect any anomalies over a year.



import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# Simulated Datasets
# -----------------------------

# Sample dataset with missing values
df_customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5],
    'Email': ['a@example.com', np.nan, 'c@example.com', np.nan, 'e@example.com']
})

df_sales = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105],
    'TransactionDate': ['2023-01-01', np.nan, '2023-01-03', '2023-01-04', np.nan]
})

df_employees = pd.DataFrame({
    'EmpID': [1, 2, 3, 4],
    'Department': ['HR', np.nan, 'IT', np.nan]
})

# Sample dataset with duplicates
df_customers_dup = pd.DataFrame({
    'CustomerID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David']
})

# Sample dataset with inconsistent formats
df_contacts = pd.DataFrame({
    'Phone': ['123-456-7890', '(123)456-7890', '1234567890', '123.456.7890'],
    'Date': ['01/02/2023', '2023-02-01', 'Feb 1, 2023', '02-01-2023'],
    'State': ['CA', 'Calif.', 'CA', 'California']
})

# Sample time series for data drift
df_revenue = pd.DataFrame({
    'Month': pd.date_range(start='2023-01-01', periods=6, freq='M'),
    'Revenue': [1000, 1050, 1020, 3000, 3100, 2950]  # Note the jump = possible drift
})

df_engagement = pd.DataFrame({
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
    'ActiveUsers': [200, 210, 205, 350]
})

df_stocks = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Price': [150, 152, 149, 148, 147, 250, 252, 251, 248, 150, 149, 151]
})

# -------------------------------------
# 1. MISSING DATA CHECKS
# -------------------------------------

print("Missing Email Records:", df_customers['Email'].isna().sum())
print("Missing Transaction Dates (%):", df_sales['TransactionDate'].isna().mean() * 100, "%")
print("Missing Department Info:", df_employees['Department'].isna().sum())

# -------------------------------------
# 2. DUPLICATE DATA CHECKS
# -------------------------------------

print("\nDuplicate Customer Records:\n", df_customers_dup.duplicated().sum())

# -------------------------------------
# 3. INCONSISTENT FORMAT CHECKS
# -------------------------------------

print("\nPhone Formats Detected:")
print(df_contacts['Phone'].apply(lambda x: 'Digits only' if x.isdigit() else 'Formatted'))

print("\nDate Format Samples:")
print(df_contacts['Date'])

print("\nState Abbreviations Unique Values:")
print(df_contacts['State'].unique())

# -------------------------------------
# 4. DATA DRIFT CHECKS
# -------------------------------------

# Monthly Revenue Drift
plt.figure(figsize=(8, 4))
sns.lineplot(data=df_revenue, x='Month', y='Revenue', marker='o')
plt.title('Monthly Revenue Drift')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Quarterly Engagement Drift
plt.figure(figsize=(6, 4))
sns.barplot(data=df_engagement, x='Quarter', y='ActiveUsers', palette='viridis')
plt.title('User Engagement per Quarter')
plt.show()

# Stock Price Anomalies
plt.figure(figsize=(8, 4))
sns.lineplot(data=df_stocks, x='Date', y='Price', marker='o')
plt.title('Stock Price Drift Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```
