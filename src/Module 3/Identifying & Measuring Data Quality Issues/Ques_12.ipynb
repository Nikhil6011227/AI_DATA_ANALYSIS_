```python
# Data Profiling Techniques Examples

# 1. Descriptive Statistics:
# Task 1: Calculate the mean, median, and mode for sales figures in a retail dataset.
# Task 2: Analyze the average age, median, and mode in a customer demographic
# dataset.
# Task 3: Determine the mean, median, and mode of daily website visit counts.






# 2. Distribution Analysis:
# Task 1: Visualize the distribution of product prices using histograms.
# Task 2: Create a distribution chart for exam scores in an educational dataset.
# Task 3: Analyze the distribution of order quantities for inventory management.






# 3. Data Type Validation:
# Task 1: Validate numeric fields in a dataset to ensure they contain only numbers.
# Task 2: Check for valid date formats in a transaction log.
# Task 3: Validate email formats in a customer contact dataset.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# ----------------------------
# Simulated Datasets
# ----------------------------

# Descriptive Statistics Dataset
df_retail = pd.DataFrame({'Sales': [250, 300, 270, 310, 295, 260, 400]})
df_customers = pd.DataFrame({'Age': [22, 35, 29, 40, 28, 30, 35, 22, 29]})
df_website = pd.DataFrame({'Visits': [150, 160, 155, 170, 165, 158, 172]})

# Distribution Analysis Dataset
df_products = pd.DataFrame({'Price': [10.5, 12.0, 9.8, 15.5, 13.0, 10.0, 11.5]})
df_scores = pd.DataFrame({'ExamScore': [78, 85, 82, 90, 88, 76, 95, 92, 85]})
df_inventory = pd.DataFrame({'OrderQty': [5, 10, 12, 7, 6, 9, 11, 10, 8]})

# Data Type Validation Dataset
df_validation = pd.DataFrame({
    'Amount': ['100', '200', 'abc', '300'],
    'TransactionDate': ['2023-01-01', '2023-02-15', '15/03/2023', '2023-04-01'],
    'Email': ['john@example.com', 'noemail@', 'jane.doe@gmail.com', 'admin@domain.org']
})

# ----------------------------
# 1. DESCRIPTIVE STATISTICS
# ----------------------------

print("Retail Sales Stats:", "\nMean:", df_retail['Sales'].mean(),
      "\nMedian:", df_retail['Sales'].median(),
      "\nMode:", df_retail['Sales'].mode()[0])

print("\nCustomer Age Stats:", "\nMean:", df_customers['Age'].mean(),
      "\nMedian:", df_customers['Age'].median(),
      "\nMode:", df_customers['Age'].mode()[0])

print("\nWebsite Visit Stats:", "\nMean:", df_website['Visits'].mean(),
      "\nMedian:", df_website['Visits'].median(),
      "\nMode:", df_website['Visits'].mode()[0])

# ----------------------------
# 2. DISTRIBUTION ANALYSIS
# ----------------------------

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.histplot(df_products['Price'], bins=5, kde=True, color='skyblue')
plt.title('Product Price Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df_scores['ExamScore'], bins=5, kde=True, color='green')
plt.title('Exam Score Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df_inventory['OrderQty'], bins=5, kde=True, color='orange')
plt.title('Order Quantity Distribution')

plt.tight_layout()
plt.show()

# ----------------------------
# 3. DATA TYPE VALIDATION
# ----------------------------

# Task 1: Validate numeric fields (Amount)
df_validation['Amount_Valid'] = df_validation['Amount'].apply(lambda x: x.isdigit())
print("\nAmount Validity:\n", df_validation[['Amount', 'Amount_Valid']])

# Task 2: Validate date format (YYYY-MM-DD)
df_validation['Date_Valid'] = df_validation['TransactionDate'].apply(
    lambda x: bool(re.match(r'^\d{4}-\d{2}-\d{2}$', x)))
print("\nDate Validity:\n", df_validation[['TransactionDate', 'Date_Valid']])

# Task 3: Validate email format
email_regex = r'^[\w\.-]+@[\w\.-]+\.\w+$'
df_validation['Email_Valid'] = df_validation['Email'].apply(
    lambda x: bool(re.match(email_regex, x)))
print("\nEmail Validity:\n", df_validation[['Email', 'Email_Valid']])




```
