```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.








# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.









# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.



import pandas as pd
from sklearn.datasets import load_iris

# Part 1: Load Iris Dataset & Check Missing Values
iris_sklearn = load_iris()
iris = pd.DataFrame(data=iris_sklearn.data, columns=iris_sklearn.feature_names)
iris['target'] = iris_sklearn.target

# Check for missing values in the entire Iris dataset
missing_values = iris.isnull().sum()
print("Missing values in Iris dataset:")
print(missing_values[missing_values > 0] if missing_values.any() else "No missing values found.")

# Part 2: Identify Inconsistent Entries in a Sample Dataset for 'Gender'
sample_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['F', 'M', 'Male', 'female', 'male']  # intentionally inconsistent
}

df_sample = pd.DataFrame(sample_data)

# Normalize the 'Gender' column to lower case for comparison
df_sample['Gender_lower'] = df_sample['Gender'].str.lower()

# Identify inconsistent entries in 'Gender' (expecting only 'male' or 'female')
valid_genders = {'male', 'female', 'm', 'f'}
inconsistent_entries = df_sample[~df_sample['Gender_lower'].isin(valid_genders)]

print("\nInconsistent 'Gender' entries:")
print(inconsistent_entries[['Name', 'Gender']])

# Optional: Standardize Gender column to consistent labels ('Male' or 'Female')
def standardize_gender(val):
    val = val.lower()
    if val in ['m', 'male']:
        return 'Male'
    elif val in ['f', 'female']:
        return 'Female'
    else:
        return 'Unknown'

df_sample['Gender_standardized'] = df_sample['Gender'].apply(standardize_gender)
print("\nStandardized Gender column:")
print(df_sample[['Name', 'Gender_standardized']])

# Part 3: Generate Data Quality Report for Iris Dataset
print("\n--- Iris Dataset Summary ---")
print(iris.describe())
```
