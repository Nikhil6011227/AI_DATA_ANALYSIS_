```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.



import matplotlib.pyplot as plt
import pandas as pd

# Sales data
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 150, 90, 120]
}

df = pd.DataFrame(data)

# Customize plot style
plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Sales'], color='orange')

# Add title and labels with custom font sizes
plt.title('Sales Comparison by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
