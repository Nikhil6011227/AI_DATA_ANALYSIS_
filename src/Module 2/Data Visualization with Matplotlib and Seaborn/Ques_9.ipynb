```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.



import matplotlib.pyplot as plt
import pandas as pd

# Data for bar plot
sales_data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 150, 90, 120]
}
df_sales = pd.DataFrame(sales_data)

# Data for scatter plot
height_weight_data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76]
}
df_hw = pd.DataFrame(height_weight_data)

# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot on the first subplot
axs[0].bar(df_sales['Product'], df_sales['Sales'], color='teal')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Product')
axs[0].set_ylabel('Sales')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot on the second subplot
axs[1].scatter(df_hw['Height'], df_hw['Weight'], color='purple')
axs[1].set_title('Height vs Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
