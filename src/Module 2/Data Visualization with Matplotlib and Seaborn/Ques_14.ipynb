```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.




import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Create line plot with custom styles
plt.plot(x, y, color='darkblue', marker='o', linestyle='--', linewidth=2, markersize=8, label='Sample Data')

# Add title and labels
plt.title('Customized Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X Axis', fontsize=14)
plt.ylabel('Y Axis', fontsize=14)

# Add grid
plt.grid(True, linestyle=':', linewidth=0.7)

# Add legend
plt.legend()

plt.show()
```
