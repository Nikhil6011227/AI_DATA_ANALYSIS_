```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.



import matplotlib.pyplot as plt

# Sample data for bar plot (product sales)
products = ['A', 'B', 'C', 'D']
sales = [25, 40, 35, 50]

# Sample data for scatter plot (height vs weight)
height = [160, 170, 165, 180, 175]
weight = [55, 65, 60, 80, 76]

# Create subplots: 1 row, 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot on first subplot
ax1.bar(products, sales, color='skyblue')
ax1.set_title('Product Sales')
ax1.set_xlabel('Product')
ax1.set_ylabel('Sales')
ax1.grid(True, linestyle='--', alpha=0.7)

# Scatter plot on second subplot
ax2.scatter(height, weight, color='orange', marker='o')
ax2.set_title('Height vs Weight')
ax2.set_xlabel('Height (cm)')
ax2.set_ylabel('Weight (kg)')
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
