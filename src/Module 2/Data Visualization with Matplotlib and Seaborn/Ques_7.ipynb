```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].




import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
}

df = pd.DataFrame(data)

# Create scatter plot with size encoding
sns.scatterplot(x='Height', y='Weight', size='Size', data=df, sizes=(50, 300), legend=False)

# Add title and labels
plt.title('Height vs Weight (with Point Sizes)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Show plot
plt.show()
```
