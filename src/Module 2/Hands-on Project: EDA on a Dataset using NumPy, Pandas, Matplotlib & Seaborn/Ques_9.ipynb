```python
# Title: Feature Engineering for Classification
# Description: Create a new feature that could help distinguish between species based on
# logical assumptions and verify its utility.



import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Feature Engineering: Create a new feature - Petal Area (Length Ã— Width)
df['petal_area'] = df['petal_length'] * df['petal_width']

# View basic statistics grouped by species
print(df.groupby('species')['petal_area'].describe())

# Visualize the new feature to verify utility
sns.boxplot(x='species', y='petal_area', data=df, palette='Set2')
plt.title('Petal Area by Species')
plt.show()
```
